```python
# Rename columns for ease of access
df = df.rename(columns={
    'S/N:': 'Sn',
    'Unnamed: 1':'Names',
    'CLASS': 'Class',
    'CLASS WORK 1 (10)': 'Classwork1',
    'ASSIGNMENT 1 (10)': 'Assignment1',
    'CLASS WORK 2 (10)': 'Classwork2',
    'ASSIGNMENT 2 (10)': 'Assignment2',
    'TOTAL (40)': 'Total40',
    'ACTUAL SCORE 1 (40/4)': 'ActualScore1',
    '1ST C.A. (10)': 'FirstCA',
    'HALF TERM SCORE (20)': 'HalfTermScore',
    'PROJECT (10)': 'Project',
    '2ND C. A. (10)': 'SecondCA',
    'CLASS WORK 4 (20)':'ClassworkTotal',
    'TOTAL (20)': 'Total20',
    'TOTAL C.A.(40)': 'TotalCA',
    'EXAM (60)': 'Exam',
    'GRAND TOTAL (100)': 'GrandTotal',
    'GRADE (A-F)': 'Grade',
    'EFFORT (1-4)': 'Effort',
    'POSITION (1ST -25TH)': 'Position',
    'COMMENT': 'Comment'
})
```

```python
# Fill missing values for numerical columns with 0 (or any other relevant approach)
df.fillna(0, inplace=True)
df
```

```python
# Define the list of values to remove
rows_to_remove = ["HIGHEST SCORE", "LOWEST SCORE", "AVERAGE", "TEACHER",  0]

# Filter out rows where 'Names' column contains any of these values
df = df[~df['Names'].isin(rows_to_remove)]

# Reset the index to keep it organized
df = df.reset_index(drop=True)
df
```

```python
df.columns = df.columns.str.strip()  # Remove any leading/trailing spaces
df = df.drop(columns=['S/N', 'Class', 'Position', 'Comment'])

df
```

```python
# Set 'Total40' column to 0
df['Total40'] = 0
df['ActualScore1'] = 0
df['HalfTermScore'] = 0

# Update 'Total40' with the sum of the specified columns
df['Total40'] = (df['Classwork1'] + df['Assignment1'] + df['Classwork2'] + df['Assignment2']).round(2)

# Divide Total by 4 to get the actual score
df['ActualScore1'] = (df['Total40'] / 4).round(2)


df['HalfTermScore'] = (df['ActualScore1'] + df['FirstCA']).round(2)
# Display the updated DataFrame
df

```

```python
# Assuming the column you want to split is named 'ClassworkTotal'
df['Classwork3'] = df['ClassworkTotal'] / 2
df['Classwork4'] = df['ClassworkTotal'] / 2

# Drop the original column
df = df.drop(columns='ClassworkTotal')
```

```python
# List the columns in the desired order
new_column_order = [
    'Names', 'Classwork1', 'Assignment1', 'Classwork2', 'Assignment2', 'Total40', 'ActualScore1',
    'FirstCA', 'HalfTermScore', 'Project', 'SecondCA', 'Classwork3', 'Classwork4', 'Total20',
    'TotalCA', 'Exam', 'GrandTotal', 'Grade', 'Effort'
]

# Reorder the DataFrame columns
df = df[new_column_order]
df
```

```python
# Display duplicate columns if any
duplicate_columns = df.columns[df.columns.duplicated()]
print("Duplicate columns:", duplicate_columns)

# Remove duplicate columns, keeping the first occurrence
df = df.loc[:, ~df.columns.duplicated()]
df
```

```python
# Add project, Classwork3 and 4, then divide by 2 and round up to 2 decimal place
# Calculate the average and round it to 2 decimal places, using .loc to avoid the warning
df.loc[:, 'Total20'] = ((df['Project'] + df['SecondCA'] + df['Classwork3'] + df['Classwork4']) / 2).round(2)
df
```

```python
# Round the specified columns to 2 decimal places
df['Classwork1'] = df['Classwork1'].round(2)
df['TotalCA'] = (df['HalfTermScore'] + df['Total20']).round(2)
df['GrandTotal'] = (df['Exam'] + df['TotalCA']).round(2)

# Define a function to assign grades based on 'GrandTotal'
def assign_grade(total):
    if total >= 75:
        return 'A1'
    elif total >= 70:
        return 'B2'
    elif total >= 65:
        return 'B3'
    elif total >= 60:
        return 'C4'
    elif total >= 55:
        return 'C5'
    elif total >= 50:
        return 'C6'
    elif total >= 45:
        return 'D7'
    elif total >= 40:
        return 'E8'
    else:
        return 'F9'

# Apply the function to the 'GrandTotal' column to create the 'Grade' column
df['Grade'] = df['GrandTotal'].apply(assign_grade)

# Display the updated DataFrame
df
```

```python
# Assuming the column you want to filter is named 'ColumnName'
df = df[(df['Classwork1'] != 0.00) & (df['Classwork1'] <= 10)]

df = df.drop(columns='Effort')
# Display the updated DataFrame
df
```
